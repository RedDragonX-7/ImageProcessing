```python
# this is a single line comment
"""this is a multi line comment
also called a docstring"""
print("-" * 30)
print("DATA TYPE AND VARIABLES")
print("-" * 30)
# numbers
age = 20 # integer
price = 19.99 # float
complex_number = 2 + 3j # complex number

# Text
name = "Alice"  # string
greeting = 'Hello, World!' #

# Boolean
is_student = True
is_employed = False

# Dispaly variables and their types
print(f"Age:{age} (Type: {type(age)})")
print(f"Price: {price} (Type: {type(price)})")
print(f"Complex Number: {complex_number} (Type: {type(complex_number)})")
print(f"Name: {name} (Type: {type(name)})")
print(f"Greeting: {greeting} (Type: {type(greeting)})")
print(f"Is Student: {is_student} (Type: {type(is_student)})")
print(f"Is Employed: {is_employed} (Type: {type(is_employed)})")

# Type conversion
print("-" * 30)
print("\n Type Conversion")
print("-" * 30)

# string to integer
num_str = "30"
num_int = int(num_str)
print(f"Converting:{num_str} to integer: {num_int} (Type: {type(num_int)})")

# integer to float
num_float = float(num_int)
print(f"Converting: {num_int} to float: {num_float} (Type: {type(num_float)})")

# number to string
num_str_again = str(num_float)
print(f"Converting: {num_float} to string: {num_str_again} (Type: {type(num_str_again)})")

# float to integer
num_int_again = int(num_float)
print(f"Converting: {num_float} to integer: {num_int_again} (Type: {type(num_int_again)})")
```

```python
# STRING OPERATIONS
print("-" * 30)
print("\n STRING OPERATIONS")
print("-" * 30)

greeting = "Hello"
name = "Bob"

# string concatenation
combined = greeting + "" +name
print(f"Concatenation: {combined}")

# string repetition
laugh = "Ha" * 3
print(f"Repetition: {laugh}")

# string methods
print("-" * 30)
text = " Python Programming "
print("-" * 30)
print(f"Original text: '{text}'")
print(f"strip whitespace: '{text.strip()}'")
print(f"Lowercase: '{text.lower()}'")
print(f"Uppercase: '{text.upper()}'")
print(f"Replace python with java: '{text.replace('Python', 'Java')}'")
print(f"Split into words: {text.split()}")


# string formatting(f-strings)
print("-" * 30)
print("\n STRING FORMATTING (f-strings)")
print("-" * 30)
name = "Charlie"
score = 95.5
print(f"student: {name}, scored {score}")

# string slicing
print("-" * 30)
print("\n STRING SLICING")
print("-" * 30)
word = "Programming"
print(f"Word: {word}")
print(f" First 3 characters: {word[:3]}")
print(f" Last 3 characters: {word[-3:]}")
print(f" Reversed: {word[::-1]}")

```

```python
# Basic operations
print( "*" *30)
print("\n BASIC OPERATIONS")
print( "*" *30)

a = 10
b = 3

# Arithmetic operations
print("-" *30)
print("\n ARITHMETIC OPERATIONS")
print("-" *30)
print(f"a:{a}, b:{b}")
print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")   
print(f"Division: {a} / {b} = {a / b}")
print(f"Modulus: {a} % {b} = {a % b}")
print(f"Exponentiation: {a} ** {b} = {a ** b}")

# Comparison operations
print("-" *30)
print("\n COMPARISON OPERATIONS")
print("-" *30)
print(f"{a}=={b}: {a==b}")
print(f"{a}!={b}: {a!=b}")
print(f"{a}>{b}: {a>b}")
print(f"{a}< {b}: {a<b}")
print(f"{a}>={b}: {a>=b}")
print(f"{a}<={b}: {a<=b}")

```
